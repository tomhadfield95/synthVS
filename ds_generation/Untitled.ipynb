```python
from scipy.stats import gamma
import matplotlib.pyplot as plt
import numpy as np

def gamma_score(x, a = 4, hydrophobic = False):

    if hydrophobic:
        return 3*gamma.pdf(np.abs(x), a)
    else:
        return 10*gamma.pdf(np.abs(x), a)
```

```python
x = np.linspace(0,10,1000)
y_polar = [gamma_score(i) for i in x]
y_hydrophobic = [gamma_score(i, hydrophobic = True) for i in x]
```

```python
plt.figure(figsize=(10, 6), dpi=500)
plt.plot(x,y_polar, label = 'Polar interaction')
plt.plot(x,y_hydrophobic, label = 'Hydrophobic interaction')
plt.legend()
plt.xlabel('Distance between ligand atom and synthetic residue')
plt.ylabel('Interaction contribution score')
plt.show()
```
