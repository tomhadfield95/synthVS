```python
from rdkit.DataStructs import TanimotoSimilarity
from rdkit.Chem import AllChem

def get_similarity(mol1, mol2):
    mol1_fp = AllChem.GetMorganFingerprint(mol1,3)
    mol2_fp = AllChem.GetMorganFingerprint(mol2,3)

    return TanimotoSimilarity(mol1_fp, mol2_fp)
```

```python
def filter_mol(m):
    too_similar = 0
    
    idx = 0
    while too_similar == 0 and idx < len(all_training_data):
        
        mm = all_training_data[idx]
        ts = get_similarity(m, mm)
        
        if ts > 0.8:
            too_similar = 1
        
        idx += 1
        
    return 1 - too_similar


pool = mp.Pool(mp.cpu_count() - 1)

filter_bin = pool.map(filter_mol, pdbbind_mols)
```
