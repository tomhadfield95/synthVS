```python
def remove_duplicates_contrib(df):
    

    df_no_duplicates = pd.DataFrame()

    for idx, row in df.iterrows():
        add = True
        for jdx, sow in df.iterrows():
            if jdx != idx and np.linalg.norm(row['ligand_pos'] - sow['ligand_pos']) < 0.05:
                add = False

        if add:
            df_no_duplicates = df_no_duplicates.append(row)

    return df_no_duplicates
```

```python
def add_distance_contrib(df):
    
    df['distance'] = [np.linalg.norm(row['ligand_pos'] - row['pharm_pos']) for idx, row in df.iterrows()]
    return df

def add_attributions_contrib(contrib_df, attribution_df, mol):
    
    attributions = []
    atom_type = []
    
    for idx, row in contrib_df.iterrows():
        for jdx, sow in attribution_df.iterrows(): 
            if np.linalg.norm(row['ligand_pos'] - np.array([sow['x'], sow['y'], sow['z']])) < 0.1:
                attributions.append(sow['attribution'])
        
        atom_type.append(mol.GetAtomWithIdx(int(row['lig_atom_idx'])).GetAtomicNum())
        
        
    contrib_df['attribution'] = attributions
    contrib_df['atom_type'] = atom_type
    
    return contrib_df
        
```

```python
def format_data(ligand, pharm, attribution_df):
    
    score, contrib_df, lig_gt, pharm_gt = assign_mol_label(ligand, pharm, hydrophobic=True, return_contrib_df=True)
    
    contrib_df_non_duplicates = remove_duplicates_contrib(contrib_df)
    contrib_df_non_duplicates = add_distance_contrib(contrib_df_non_duplicates)
    contrib_df_non_duplicates = add_attributions_contrib(contrib_df_non_duplicates, attribution_df, ligand)
    
    return contrib_df_non_duplicates

```

```python
root_dir_test = '/data/hookbill/hadfield/syntheticVS/data/pdbbind_filtered_test_set_hydrophobic'

all_lig_paths = sorted(glob.glob(f'{root_dir_test}/sdf/ligands/lig*.sdf'))
all_pharm_paths = sorted(glob.glob(f'{root_dir_test}/sdf/pharmacophores/pharm*.sdf'))

all_df = pd.DataFrame()

for idx in range(len(all_lig_paths)):

    lig = Chem.MolFromMolFile(all_lig_paths[idx])
    pharm = Chem.MolFromMolFile(all_pharm_paths[idx])
    attribution_df = pd.read_csv(f'/data/hookbill/hadfield/syntheticVS/data/pdbbind_filtered_test_set_hydrophobic/attribution_dfs/df{idx}.csv', sep = ' ')

    out_df = format_data(lig, pharm, attribution_df)
    
    
    if out_df.shape[0] > 0:
        out_df['example'] = [idx]*out_df.shape[0]
        all_df = all_df.append(out_df)
        

    
```
