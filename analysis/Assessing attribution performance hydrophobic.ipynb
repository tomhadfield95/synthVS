```python
all_dfs = glob.glob('/data/hookbill/hadfield/syntheticVS/data/pdbbind_filtered_test_set_hydrophobic/attribution_dfs/df*.csv')
spearman_corr = []

identified_strong_contribution = []
num_strong_contributions = []

for df_path in all_dfs:
    
    df = pd.read_csv(df_path, sep =  ' ')
    spearman_corr.append(spearmanr(df['attribution'], df['binding']).correlation)
    
    threshold = 2.5
    
    num_strong_contributions.append(df.loc[df['binding'] > threshold].shape[0])
    
    for idx, row in df.iterrows():
        if idx == 0:
            if row['binding'] > threshold: 
                identified_strong_contribution.append(1)
            else:
                identified_strong_contribution.append(0)
        
    
```

```python
spearman_corr
```

```python
spearman_corr = [x for x in  spearman_corr if pd.notnull(x)]
```

```python
plt.hist(spearman_corr)
plt.show()
```
